# 出力関数

```python
def output(posx, posy, plus_pos, ob_id):
    print('    <poly id="poly_'+str(ob_id) + '" type="building.yes" color="blue" fill="1" layer="-1.00" shape="'+str(posx) + '.' + str(random.randint(100000, 999999)) +','+ str(posy)+ '.' + str(random.randint(100000, 999999)) + ' ' +str(posx) + '.' + str(random.randint(100000, 999999)) +',' + str(posy + plus_pos)+ '.' + str(random.randint(100000, 999999)) + ' ' +str(posx + plus_pos) + '.' + str(random.randint(100000, 999999)) +',' + str(posy + plus_pos)+ '.' + str(random.randint(100000, 999999)) + ' ' +str(posx + plus_pos) + '.' + str(random.randint(100000, 999999)) +',' + str(posy)+ '.' + str(random.randint(100000, 999999)) + '"/>' )
```

# 1grid内描画関数(1回呼び出されると36回描画)


```python
def in_grid_draw(posx, posy, ob_id): #pos = grid内の１番左上の座標  ob_id= 左上のgrid_id
    column_count = 1 # 列
    for ob_index in range (1, ob_num +1): #grid内のobstacle ごとのroop
        if column_count == 6:
            output(posx,posy,plus_pos, ob_id)
            column_count = 1
            posx -=  150
            posy += 30
        else:
            output(posx,posy,plus_pos, ob_id)
            column_count += 1
            posx += 30
        ob_id += 1
```

```python
posx = 10
posy = 10
grid_count = 1
ob_count = 1
ob_id = 0 #建物のID
column_count = 1 # 列
row_count = 1 #行

for grid_index in range (1, grid_num + 1): #grid ごとのroop
#     print('-----------------------------------------------------------position posx', posx, 'posy', posy)
    if column_count == 5:
        in_grid_draw(posx,posy,ob_id) # 描画開始
        posx -= 800
        posy += 200
        column_count = 1
    else:
#         print('posx', posx, 'posy', posy)
        in_grid_draw(posx,posy,ob_id) # 描画開始
        posx += 200
        column_count += 1
    ob_id = ob_id + 36
    grid_count += 1
```
