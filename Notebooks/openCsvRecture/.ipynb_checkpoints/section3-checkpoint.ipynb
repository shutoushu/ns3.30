```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("data/src/Lena.jpg")
cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
color_list = ["blue", "green", "red"]
for i,j in enumerate(color_list):
    hist = cv2.calcHist([img],[i],None, [256], [0,256])
    plt.plot(hist, color=j)
```

```python
img_grey = cv2.imread("data/src/Lena.jpg", 0)
hist2 = cv2.calcHist([img_grey], [0],None,  [256], [0,256])
plt.plot(hist2)
```

# ２値化

```python
img = cv2.imread("data/src/grapes.jpg", 0)
cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
threshold = 100
ret, img_th = cv2.threshold(img, threshold,  255, cv2.THRESH_BINARY)
ret
```

```python
cv2.imshow("img_th", img_th)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret2, img_o = cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)
ret2
```

```python
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
plt.plot(hist)
```

```python
cv2.imshow("otsu", img_o)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_ada = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 3,1)
```

```python
cv2.imshow("ada", img_ada)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# アファイン変換

```python
img = cv2.imread("data/src/grapes.jpg")
h, w = img.shape[:2]
dx, dy = 30, 30
```

```python
afn_mat = np.float32([[1,0,dx],[0,1,dy]])
img_afn = cv2.warpAffine(img, afn_mat, (w,h))
```

```python
cv2.imshow("trans", img_afn)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
rot_mat = cv2.getRotationMatrix2D((w/2, h/2), 40, 1)
img_afn2 = cv2.warpAffine(img, rot_mat, (w,h))
cv2.imshow("rot", img_afn2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
